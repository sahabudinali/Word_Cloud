```python
from wordcloud import WordCloud,ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
text=" "

with open('gndec.txt', encoding ='utf-8') as f:
    text=''.join(f.readlines())
```

```python
wc=WordCloud()
```

```python
wc.generate(text)
```

```python
wc.to_file('output.png')
```

```python

plt.figure(figsize=(10,10))
plt.imshow(wc,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python

plt.figure(figsize=(10,10))
wc=WordCloud(max_font_size=40).generate(text)
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
custom_mask=np.array(Image.open("download.jpeg"))
wc=WordCloud(background_color="white",mask=custom_mask)
wc.generate(text)
plt.figure(figsize=(6,6))
print("               Turbon Shape")
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wc=WordCloud(background_color="white",mask=custom_mask, contour_width=3,contour_color="steelblue")
wc.generate(text)
plt.figure(figsize=(6,6))
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
custom_mask=np.array(Image.open('images.jpeg'))
wc=WordCloud(background_color="black",mask=custom_mask)
wc.generate(text)


plt.figure(figsize=(10,10))
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```
